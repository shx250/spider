# Baidu*

```python
import requests
import pandas as pd
url = 'https://top.baidu.com/api/board?platform=wise&tab=realtime'
header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36',
    'Host': 'top.baidu.com',
    'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'zh-CN,zh;q=0.9',
    'Referer': 'https://top.baidu.com/board?platform=pc&sa=pcindex_a_right'
}
r = requests.get(url,header)
print(r.status_code)
json_data = r.json()
title_list = []
order_list = []
score_list = []
desc_list = []
url_list = []
top_content_list = json_data['data']['cards'][0]['topContent']
for item in top_content_list:
    #print(item['query'])
    title_list.append(item['query'])
    order_list.append(item['index'])
    score_list.append(item['desc'])
    url_list.append(item['url'])
content_list = json_data['data']['cards'][0]['content']
for item in content_list:
    #print(item['query'])
    title_list.append(item['query'])
    order_list.append(item['index']+1)
    score_list.append(item['hotScore'])
    desc_list.append(item['desc'])
    url_list.append(item['url'])

nums=min(len(title_list),len(order_list),len(score_list),len(desc_list),len(url_list))
title_list=title_list[0:nums]
order_list=order_list[0:nums]
score_list=score_list[0:nums]
desc_list= desc_list[0:nums]
url_list=url_list[0:nums]

df = pd.DataFrame(
    {
        'title':title_list,
        'order':order_list,
        'score':score_list,
        'desc':desc_list,
        'url':url_list
    }
    )
df
```
